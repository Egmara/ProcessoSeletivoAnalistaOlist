# Olhando detalhadamente o Cliente A (para treinar como ler os dados)

```python
idx = dataCustomers[dataCustomers['customer_unique_id'] == A].index
print(idx)
dataCustomers.loc[idx,:]
```

```python
C_A = dataCustomers.customer_id[idx]
type(C_A)
```

```python
id_customer_A = np.array(C_A)
id_customer_A
```

```python
# vamos ver esses pedidos no dataset de Pedidos

idx_Pedidos_A = []
for i in range(0,17):
    idx_Pedidos_A.append(dataOrders[dataOrders.customer_id == id_customer_A[i]].index[0])
idx_Pedidos_A 
```

```python
Estado_dos_Pedidos_A = dataOrders.loc[idx_Pedidos_A,:]
Estado_dos_Pedidos_A
```

```python
order_id_A = np.array(Estado_dos_Pedidos_A.order_id)
order_id_A
```

```python
# vamos ver como foi o pagamento?
idx_Pedidos_A_pay = []
for i in range(0,17):
    idx_Pedidos_A_pay.append(dataPay[dataPay.order_id == order_id_A[i]].index[0])
idx_Pedidos_A_pay
```

```python
Pay_dos_Pedidos_A = dataPay.loc[idx_Pedidos_A_pay,:]
Pay_dos_Pedidos_A
```

```python
# vamos ver as reviews?

idx_Pedidos_A_rev = []
for i in range(0,17):
    idx_Pedidos_A_rev.append(dataRev[dataRev.order_id == order_id_A[i]].index[0])
idx_Pedidos_A_rev
```

```python
Rev_dos_Pedidos_A = dataRev.loc[idx_Pedidos_A_rev,:]
Rev_dos_Pedidos_A
```

```python
# vamos ver os items?

idx_Pedidos_A_item = []
for i in range(0,17):
    aux = dataItems[dataItems.order_id == order_id_A[i]].index
    if aux.shape[0] != 0:
        idx_Pedidos_A_item.append(aux[0])
idx_Pedidos_A_item
```

```python
I_dos_Pedidos_A = dataItems.loc[idx_Pedidos_A_item,:]
I_dos_Pedidos_A
```

```python
# vamos ver os produtos?

prod_id_A = np.array(I_dos_Pedidos_A.product_id)
prod_id_A 

idx_Pedidos_A_prod = []
for i in range(0,16):
    aux = dataProd[dataProd.product_id == prod_id_A[i]].index
    if aux.shape[0] != 0:
        idx_Pedidos_A_prod.append(aux[0])
idx_Pedidos_A_prod
```

```python
Prod_dos_Pedidos_A = dataProd.loc[idx_Pedidos_A_prod,:]
Prod_dos_Pedidos_A
```

```python
# quem são os sellers que esse cliente mais usa?

seller_id_A = np.array(I_dos_Pedidos_A.seller_id)

idx_Pedidos_A_s = []
for i in range(0,16):
    aux = dataSellers[dataSellers.seller_id == seller_id_A[i]].index
    if aux.shape[0] != 0:
        idx_Pedidos_A_s.append(aux[0])
idx_Pedidos_A_s
```

```python
Seller_dos_Pedidos_A = dataSellers.loc[idx_Pedidos_A_s,:]
Seller_dos_Pedidos_A
```
